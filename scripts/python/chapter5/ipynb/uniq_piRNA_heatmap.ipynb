```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Set the background to white
sns.set(style="white")


# File path to your data
file_path = '/mnt/home3/miska/nm667/scratch/inProgress/mice_PiRNA/results/STAR_uniqe_marker_strain_wise/seqCount/all_uniq_piRNA_count_for_figure.tab'  # Replace with your file path

# Read the TSV file
data = pd.read_csv(file_path, sep='\t', header=None)
# Randomly sample 1000 rows from the data
# Define column names
data.columns = ['Sequence Name', 'Count', 'Strain', 'Replicate', 'Timepoint']

sampled_data = data.sample(n=100000, random_state=1)

# Strains and timepoints order
strains_order = ['C57BL_6NJ', 'BALB_cJ', 'A_J', 'FVB_NJ', 'C3H_HeJ', '129S1_SvImJ', 'DBA_2J', 'AKR_J', 'CBA_J', 'NOD_ShiLtJ', 'WSB_EiJ', 'CAST_EiJ', 'PWK_PhJ', 'SPRET_EiJ']
timepoints_order = ['E16.5', 'P12.5', 'P20.5']

# Averaging over replicates in the sampled data
avg_data = sampled_data.groupby(['Timepoint', 'Strain', 'Sequence Name']).agg({'Count': 'mean'}).reset_index()

# Pivot the data for heatmap
heatmap_data = avg_data.pivot_table(index='Sequence Name', columns=['Timepoint', 'Strain'], values='Count', aggfunc='mean')

# Function to order sequence names by count values for a specific strain
def order_sequences_by_strain(data, strain):
    strain_data = data[data['Strain'] == strain]
    ordered_indices = strain_data.groupby('Sequence Name')['Count'].mean().sort_values(ascending=False).index
    return ordered_indices

# Apply the ordering for each strain and store the order
sequence_order_per_strain = {}
for strain in strains_order:
    sequence_order_per_strain[strain] = order_sequences_by_strain(avg_data, strain)

# Combine the ordered sequence names for all strains
combined_sequence_order = pd.concat([pd.Series(sequence_order_per_strain[strain]) for strain in strains_order]).unique()

# Reorder the heatmap data according to the combined sequence order
reordered_heatmap_data = heatmap_data.reindex(combined_sequence_order)

# Plotting the reordered heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(reordered_heatmap_data, cmap='viridis', cbar_kws={'label': 'Count'})
plt.title('Heatmap of Count Values (Sampled and Reordered)')
plt.xticks(rotation=90, ha='center')  # Rotate x-ticks to 90 degrees
plt.yticks([])  # Hides sequence name labels
plt.xlabel('')  # Clear the default xlabel


# Move the x-axis to the top
ax = plt.gca()
ax.xaxis.tick_top()
plt.xticks(rotation=90, ha='center') 
ax.xaxis.set_label_position('top')

# Save the plot as a PDF
pdf_file_path = '/mnt/home3/miska/nm667/scratch/inProgress/mice_PiRNA/results/STAR_uniqe_marker_strain_wise/seqCount/heatmap_output.pdf'
plt.savefig(pdf_file_path, format='pdf', bbox_inches='tight')

plt.show()
```
