```r
library(foreach)
library(doParallel)


# Create a directory to save the plots
if (!dir.exists("volcano_plots")) {
    dir.create("volcano_plots")
}


# Set up parallel processing
num_cores <- 144 # Set the number of CPU cores to use
registerDoParallel(cores = num_cores)

# Define a function for generating a volcano plot
generate_volcano_plot <- function(col1, col2, cds) {
    contrast <- c('sample',col1, col2)
    res <- results(cds, contrast = contrast)

    # Create a color vector based on log2FoldChange and p-value
    colors <- ifelse(res$pvalue <= 0.05 & abs(res$log2FoldChange) >= 2, "red",
                             ifelse(res$pvalue <= 0.05, "blue", "gray"))
    
    volcano_plot <- EnhancedVolcano(res,
                                        lab = rownames(res),
                                        x = 'log2FoldChange',
                                        y = 'pvalue',
                                        title = paste(col1, "vs", col2, "Volcano Plot"),
                                        pCutoff = 0.05,
                                        FCcutoff = 2,
                                        col=colors,  #c('blue', 'blue', 'blue', 'red3'),
                                        colAlpha = 1,    cutoffLineType = 'blank',
                                        cutoffLineCol = 'black',
                                        cutoffLineWidth = 0.8,
                                        hline = c(10e-12, 10e-36, 10e-60, 10e-84),
                                        hlineCol = c('grey0', 'grey25','grey50','grey75'),
                                        hlineType = 'longdash',
                                        hlineWidth = 0.8,
                                        gridlines.major = FALSE,
                                        gridlines.minor = FALSE,
                                        xlim = c(-10, 10))

    # Add text labels for upregulated, downregulated, and nonsignificant genes
    textLabels <- ifelse(res$pvalue <= 0.05 & abs(res$log2FoldChange) >= 2,
                                 ifelse(res$log2FoldChange > 0, "Up", "Down"), "NS")
            
    text(x = res$log2FoldChange, y = -log10(res$pvalue), labels = textLabels, cex = 0.7, col = "black")
            
    
    plot_filename <- paste("volcano_plots/", col1, "_vs_", col2, "_volcano_plot.png", sep = "")
    png(plot_filename, width = 1000, height = 1200)
    print(volcano_plot)
    dev.off()
}


# Parallelize volcano plot generation using foreach
foreach(i = 1:(length(column_names) - 1),
        .combine = 'c') %dopar% {
    for (j in (i + 1):length(column_names)) {
        col1 <- column_names[i]
        col2 <- column_names[j]
        
        # Check if col1 and col2 have the same name
        if (col1 != col2) {
            generate_volcano_plot(col1, col2, cds)
        }
    }
}

# Stop parallel processing
stopImplicitCluster()
```
