```r
plotPCA.san <- function (object, intgroup = "condition", ntop = 20000, returnData = FALSE, pdf_filename = "PCA_plots.pdf") {
  pdf(pdf_filename)
  
  rv <- rowVars(assay(object))
  select <- order(rv, decreasing = TRUE)[seq_len(min(ntop, length(rv)))]
  pca <- prcomp(t(assay(object)[select, ]))
  percentVar <- pca$sdev^2/sum(pca$sdev^2)
  
  if (!all(intgroup %in% names(colData(object)))) {
    stop("the argument 'intgroup' should specify columns of colData(dds)")
  }
  
  intgroup.df <- as.data.frame(colData(object)[, intgroup, drop = FALSE])
  group <- if (length(intgroup) > 1) {
    factor(apply(intgroup.df, 1, paste, collapse = " : "))
  } else {
    colData(object)[[intgroup]]
  }
  
  num_pcs <- min(10, ncol(pca$x))  # To handle cases with fewer than 10 PCs
  
  d <- data.frame(matrix(pca$x[, 1:num_pcs], ncol = num_pcs), Samples = group, intgroup.df, name = colData(rl)[,1])
  
  if (returnData) {
    attr(d, "percentVar") <- percentVar[1:num_pcs]
    return(d)
  }
  
  # Loop to create scatter plots for PC1 to PC10 against each other
  pc_plots <- list()
  for (i in 1:num_pcs) {
    for (j in (i + 1):num_pcs) {

        # Create a new column 'Shape' based on the sample names
      pc_plots[[length(pc_plots) + 1]] <- ggplot(data = d, aes_string(x = paste0("PC", i), y = paste0("PC", j), color = "Samples", label = "name")) +
        geom_point(size = 1.5, position = position_jitter(width = 1, height = 1)) +
        xlab(paste0("PC", i, ": ", round(percentVar[i] * 100), "% variance")) +
        ylab(paste0("PC", j, ": ", round(percentVar[j] * 100), "% variance")) +
        scale_color_manual(values = final_colors) 
        theme(legend.text = element_text(size = 5)) +
        facet_wrap(~Samples, scales = "free") +
        coord_fixed() +
        theme_minimal()
    }
  }
  #pdf(paste0("all_timpoint_allPC_10_.pdf"),paper="a4")
  # Print all PC vs PC scatter plots and save to PDF
  for (i in 1:length(pc_plots)) {
    #pdf(paste0("all_timpoint_all_PC",i,".pdf"),paper="a4")
    print(pc_plots[[i]])
    #dev.off()
  }
  
  # Bar plot for eigenvalues of PC1 to PC10
  eigenvalue_data <- data.frame(Component = paste0("PC", 1:num_pcs), Eigenvalue = percentVar[1:num_pcs])
  eigenvalue_plot <- ggplot(eigenvalue_data, aes(x = Component, y = Eigenvalue)) +
    geom_bar(stat = "identity", fill = "blue", alpha = 0.7) +
    xlab("Principal Component") +
    ylab("Eigenvalue (%)") +
    ggtitle("Eigenvalues of Principal Components (PC1 to PC10)") +
    theme_minimal()

  # Print the eigenvalue bar plot and save to PDF
  #pdf(paste0("all_timepoint_eigenvectore_bar_PC.pdf"),paper="a4")
  print(eigenvalue_plot)
  
  # Close the PDF device
  dev.off()
}

plotPCA.san(rl, intgroup = "sample", ntop = nrow(counts(cds)))
```
