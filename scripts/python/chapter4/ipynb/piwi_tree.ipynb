```r
library(bio3d)
library(phangorn)

# Read the FASTA file with protein sequences
fasta_file <- "/mnt/home3/miska/nm667/scratch/inProgress/mice_PiRNA/analysis/RNA_deSeq/piwi_protiens_seq/PIWIL1.fa"
sequences <- read.fasta(fasta_file)

# Perform multiple sequence alignment using muscle
alignment <- muscle(sequences)

# Convert the alignment to a matrix
alignment_matrix <- msaConvert(alignment, type = "matrix")

# Construct a distance matrix based on alignment
distance_matrix <- dist.alignment(alignment_matrix)

# Construct a tree using Neighbor-Joining method
tree <- nj(distance_matrix)

# Plot the tree
plot(tree)
```
